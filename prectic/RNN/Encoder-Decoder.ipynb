# Encoder-Decoder
The encoder-decoder architecture is a neural network design used for sequence-to-sequence (Seq2Seq) tasks in deep learning. It is widely applied in areas such as machine translation, text summarization, image captioning, and more.

# Encoder

1. The encoder processes the input sequence and compresses it into a fixed-size context vector (or a sequence of hidden states in some variants).
2. Typically, it uses architectures like RNNs, LSTMs, GRUs to capture temporal dependencies and relationships in the input sequence.
3. For example, in machine translation, the input could be a sentence in one language.

# Context/Latent Representation

1. This is the intermediate representation output by the encoder.
2. It acts as a summary of the input sequence and is used by the decoder to generate the output.

# Decoder

1. The decoder generates the output sequence step-by-step using the context vector and previously generated outputs (auto-regressive decoding).
2. Like the encoder, it often uses RNNs, LSTMs, GRUs, or Transformers.
3. For instance, in machine translation, the output could be a sentence in another language.

# How It Works (Step-by-Step)
1. <b>Encoding Phase</b>:

The encoder takes the input sequence and processes it through its layers to generate the context vector (or sequence of hidden states).

2. <b>Decoding Phase</b>:

The decoder starts generating the output sequence, one token at a time.
At each step, the decoder considers the context vector and its previously generated tokens to predict the next token.

# Applications
1. <b>Machine Translation</b>: Translate a sentence from one language to another.
2. <b>Image Captioning</b>: Generate textual descriptions of images (Encoder: CNN, Decoder: RNN/Transformer).
3. <b>Speech Recognition</b>: Convert audio sequences into text.

# Teacher Forcing
Teacher Forcing is a training technique used in sequence-to-sequence models, particularly in encoder-decoder architectures. It is applied during the training phase to improve the model's convergence and performance.

1. In the decoder of a sequence-to-sequence model, predictions are generated sequentially, one token at a time.
2. Without Teacher Forcing: At each time step, the decoder uses its previous prediction as input for the next step.
3. With Teacher Forcing: Instead of using the model's own predictions as input for the next step, the ground truth (actual target sequence) is used.

# Challenges
1. <b>Long Sequence Handling:</b> The encoder compresses the entire input sequence into a single fixed-size context vector. For long sequences, this can lead to information loss, as the context vector struggles to capture all the details.

# Attention Mechanism
The attention mechanism is a powerful concept in deep learning, particularly in sequence-to-sequence models like the encoder-decoder architecture. It allows the model to focus on specific parts of the input sequence while generating each output token, addressing challenges like long sequence handling and vanishing gradients.

1. At each decoding step, the decoder decides which parts of the encoder's output (hidden states) are most relevant.
2. Instead of relying on a single fixed context vector (as in the vanilla encoder-decoder), attention computes a weighted sum of all encoder hidden states.
3. The weights (or attention scores) are dynamically computed based on the decoder's current state and each encoder hidden state.

# Working of Attention Mechanism
The attention mechanism addresses the limitation of the encoder-decoder architecture by allowing the decoder to selectively focus on specific parts of the input sequence when generating each output token. Instead of relying solely on a fixed-length context vector, attention computes a dynamic context vector at each decoding step.

# Step-by-Step Explanation
1. Inputs to the Attention Mechanism

       Encoder hidden states (): These are outputs from the encoder for each token in the input sequence.
       Decoder hidden state at time (): The current hidden state of the decoder, which represents the state of decoding at step .

2. Compute Alignment Scores: The goal is to calculate how well each encoder hidden state aligns with the current decoder state. This is done using a scoring function.

        Scoring Functions: The scoring function score(,)computes a scalar alignment score for the i-th encoder hidden state and the current decoder state . The attention mechanism often uses a neural network (ANN) to calculate the scoring function







# Additive (Bahdanau) Attention
In additive attention, the alignment score , is computed using a small neural network with learnable parameters. The process is as follows:

1. Concatenate the decoder hidden state  and the encoder hidden state .
2. Pass the concatenated vector through a feedforward neural network with a single hidden layer and a non-linear activation (typically tanh).
3. Use a weight vector (learned during training) to produce the final scalar score.

 



4. Why Use ANN for Scoring: universal function approximator

# Luong Attention
Unlike Bahdanau (additive) attention, Luong attention uses multiplicative (dot-product) scoring to calculate alignment scores. It's computationally simpler and often faster to compute.
