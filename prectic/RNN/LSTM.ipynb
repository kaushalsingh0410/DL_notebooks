# LSTM (Long Short-Term Memory)
Long short-term memory (LSTM) isa type of recurrent neural network (RNN) that's designed to learn long-term dependencies and solve sequence learning problems.

LSTM architecture designed to model sequential data while addressing the vanishing gradient problem common in traditional RNNs.

## Key Features of LSTM:
1. <b> Memory Cell </b> : Maintains information over long time periods.
2. <b> Forget Gate </b>: Decides what information to discard.
3. <b> Input Gate</b>: Decides what new information to store in the memory.
4. <b> Output Gate</b>: Decides what part of the memory to output.

## Common Applications of LSTM:
1. <b>Time-Series Forecasting</b>: Stock prices, weather predictions, etc.
2. <b>Natural Language Processing (NLP)</b>: Sentiment analysis, text generation, etc.
3. <b>Speech Recognition</b>: Converting audio to text.

## Why use LSTM insted of RNN

## 1. Vanishing Gradient Problem
1. <b> RNNs</b>: Struggle with learning long-term dependencies because gradients diminish as they backpropagate through many layers (time steps). This makes it hard to update weights effectively for earlier layers in the sequence.
2. <b>LSTMs</b>: Use a memory cell and gates (forget, input, and output) to regulate the flow of information, which allows them to retain information over long sequences and mitigate vanishing gradients.

## 2. Learning Long-Term Dependencies
1. <b>RNNs</b>: Tend to remember only recent information because of the above limitation.
2. <b>LSTMs</b>: Are explicitly designed to store information for longer durations and decide what to keep or discard.

## 3. Stagnating training in RNNs
Stagnating training in RNNs occurs primarily due to the vanishing gradient problem, where gradients become extremely small as they propagate backward through time, resulting in negligible weight updates. This prevents the RNN from learning effectively, especially for long sequence

# Long-Term Memory (LTM)
1. Represented by the Cell State (𝐶𝑡) in an LSTM.
2. Captures and stores long-term dependencies across the sequence, allowing the network to retain information over many time steps.
3. Information in the cell state is modified through forget gates (to discard irrelevant information) and input gates (to add relevant new information).

# Short-Term Memory (STM)
1. Represented by the Hidden State (ℎ𝑡) in an LSTM.
2. Stores immediate, contextual information relevant to the current or recent time steps in the sequence.


# Embedding Layer
An embedding layer is a type of neural network layer that is often used with sequential models, including LSTMs, for tasks involving categorical data, such as text processing. It maps discrete items (like words or tokens) into continuous vector spaces of fixed size, where semantically similar items are closer together.

1. Input:A sequence of categorical indices, such as word indices from a vocabulary.
2. Output:A sequence of dense vectors corresponding to each input token.
