# Tanh Activation Function

#
1. Data Hidden complex pattern decode
2. Output or at Hidden
3. Mostly at Hidden Layer
4. values range b/w -1 to 1


```python
# 3:- AI:- MATHS
```

```python
output_data = np.linspace(-10,10,50)
output_data
```

```python
nr = e**output_data - e**(-output_data)
```

```python
dr = e**output_data + e**(-output_data)
```

```python
tanh = nr/dr
```

```python
tanh
```

```python
formulae = '''e**x - e**(-x)
______________
e**x + e**(-x)'''
```

```python
plt.plot(output_data,tanh)
plt.annotate(text = formulae,xy = (-5,0.75))
plt.axhline(0)
plt.show()
```

# Leaky Relu

```python
# summation(z):- {z>= 0->>> z}
#                {z<0-->> 0.01*z}
```

```python
e = np.e
```

```python
z = -3

0.01*(e**z - 1)
```

```python
z = np.linspace(-1,1,30)
```

```python
elu_act = 0.01*(e**z -1)
```

```python
elu_act
```

```python
plt.plot(z,elu_act)
plt.axhline(0)
plt.yticks(np.linspace(0.006,0.17,30))
plt.show()
```

```python
z = np.linspace(-10,10,30)

elu_act = 0.01*(e**z -1)
```

```python
plt.plot(z,elu_act)
plt.axhline(0)
```

```python
# relu
```

```python
z
```

```python
relu_act = []
for i in z:
    relu_act.append(max(i,0))
    
relu_act
        
```

```python
plt.plot(z,relu_act)
plt.axhline(0,linewidth = 0.5)
plt.axvline(0)
```

```python
lk_relu_act = []
alpha = 0.05
for i in z:
    if i>=0:
        lk_relu_act.append(i)
    else:
        lk_relu_act.append(alpha*i)
    
lk_relu_act     
```

```python
plt.plot(z,lk_relu_act)
plt.axhline(0,linewidth = 0.5)
plt.axvline(0)
```
