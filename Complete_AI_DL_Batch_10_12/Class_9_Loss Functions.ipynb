```python
y = np.random.random(10)
y
```

```python
y_hat = 1/(1+np.e**(-y))
y_hat
```

```python
log_loss = (-(y*np.log(y_hat)  + (1-y)*np.log(1-y_hat)).sum())/len(y)
```

```python
y
```

```python
dr = np.e**y
```

```python
all_nr = []
for i in y:
    all_nr.append(np.e**i)
```

```python
all_nr
```

```python
dr
```

```python
softmax_y_hat = all_nr/sum(dr)
```

```python
-(y*np.log(softmax_y_hat)).sum()
```

```python
MCSCCE = (-(y*np.log(softmax_y_hat)).sum())/len(y)
```
