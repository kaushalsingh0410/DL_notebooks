# Tanh Activation Function

#
1. Data Hidden complex pattern decode
2. Output or at Hidden
3. Mostly at Hidden Layer
4. values range b/w -1 to 1


```python
# 3:- AI:- MATHS
```

```python
output_data = np.linspace(-10,10,50)
output_data
```

```python
nr = e**output_data - e**(-output_data)
```

```python
dr = e**output_data + e**(-output_data)
```

```python
tanh = nr/dr
```

```python
tanh
```

```python
formulae = '''e**x - e**(-x)
______________
e**x + e**(-x)'''
```

```python
plt.plot(output_data,tanh)
plt.annotate(text = formulae,xy = (-5,0.75))
plt.axhline(0)
plt.show()
```
