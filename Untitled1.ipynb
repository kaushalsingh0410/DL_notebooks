```python
classes1 = np.array(classes1)
train_images1 = np.array(train_images1)
labels1 = np.array(labels1)
test_images1 = np.array(test_images1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(train_images1,labels1,test_size = 0.23,random_state = 32001)
# print(X_train.shape)
# print(X_test.shape)
# print(y_train.shape)
# print(y_test.shape)
```

```python
earlyStop = EarlyStopping(monitor = 'val_loss',
                          patience = 4,
                          min_delta = 0.0001,
                          verbose = 1)
```

```python
model1 = Sequential([
    Resizing(150,150),
    RandomContrast(0.1),
    RandomFlip(mode = 'horizontal_and_vertical'),
    RandomRotation(0.1),

    Conv2D(60,(4,4),padding = 'same',activation = 'relu'),
    MaxPool2D(3,3),
    Dropout(0.2),

    Conv2D(80,(4,4),padding = 'same',activation = 'relu'),
    MaxPool2D(3,3),
    Dropout(0.2),

    Conv2D(80,(4,4),padding = 'same',activation = 'relu'),
    MaxPool2D(3,3),
    Dropout(0.2),

    Conv2D(60,(4,4),padding = 'same',activation = 'relu'),
    MaxPool2D(3,3),
    Dropout(0.2),

    Flatten(),
    Dense(30,activation =  'relu'),
    Dense(80,activation =  'relu'),
    Dense(30,activation =  'relu'),
    Dropout(0.1),
    Dense(5,activation =  'softmax')
])
model1.compile(optimizer = 'adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
model1.summary()
```

```python
history1 = model1.fit(X_train,y_train,epochs = 50,validation_data = (X_test,y_test),callbacks = [earlyStop])
```
